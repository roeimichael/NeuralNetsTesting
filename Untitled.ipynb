```python
results = pd.read_csv("intermediate_results.csv")
mask = (results['Avg_ROI'] > 0.005) & (results['Precision'] > 0.2) & (results['total_positions'] >= 3) & (results['total_positions'] <= 350)

filtered_df = results[mask]
results_sorted = filtered_df.sort_values(by=['Avg_ROI', 'Precision','Recall', 'TP_positions'], ascending=False)
```

```python

top_models_list = []

for idx, row in results_sorted.iterrows():
    model_name = row['Model']
    model_params = row['Model_Params']  # assuming these are stored as dictionary
    model_optimizer = row['Optimizer']
    top_models_list.append({"name": model_name, "params": model_params,"optimizer": model_optimizer})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# Set the style of seaborn
sns.set_theme(style="whitegrid")

# Create subplots
fig, ax = plt.subplots(3, 2, figsize=(20, 20))

# Plot 1: Barplot of Avg_ROI by Model
sns.barplot(x="Model", y="Avg_ROI", data=results_sorted, palette="Blues_d", ax=ax[0, 0])
ax[0, 0].set_title("Average ROI by Model", fontsize=16)
ax[0, 0].set_xlabel("Model", fontsize=14)
ax[0, 0].set_ylabel("Average ROI", fontsize=14)

# Plot 2: Scatter plot of Precision vs Recall
sns.scatterplot(x="Precision", y="Recall", hue="Model", data=results_sorted, s=100, ax=ax[0, 1])
ax[0, 1].set_title("Precision vs Recall by Model", fontsize=16)
ax[0, 1].set_xlabel("Precision", fontsize=14)
ax[0, 1].set_ylabel("Recall", fontsize=14)
ax[0, 1].set_xlim(0, 1)
ax[0, 1].set_ylim(0, 1)

# 3. Histogram of 'total_positions'
sns.histplot(data=results_sorted, x="total_positions", bins=10, kde=True, ax=ax[1, 0])
ax[1, 0].set_title('Histogram of Total Positions')

# Plot 4: Barplot of Avg_ROI by Cost_Matrix_Value
sns.barplot(x="Cost_Matrix_Value", y="Avg_ROI", data=results_sorted, palette="Blues_d", ax=ax[1, 1])
ax[1, 1].set_title("Average ROI by Cost Matrix Value", fontsize=16)
ax[1, 1].set_xlabel("Cost Matrix Value", fontsize=14)
ax[1, 1].set_ylabel("Average ROI", fontsize=14)

# Plot 5: Scatter plot of TP_positions vs positive_positions
sns.scatterplot(x="positive_positions", y="TP_positions", hue="Model", data=results_sorted, s=100, ax=ax[2, 0])
ax[2, 0].set_title("TP_positions vs Positive Positions by Model", fontsize=16)
ax[2, 0].set_xlabel("Positive Positions", fontsize=14)
ax[2, 0].set_ylabel("TP Positions", fontsize=14)
positive_positions_max = results_sorted["positive_positions"].max()
ax[2, 0].set_xlim(0, positive_positions_max)
ax[2, 0].set_ylim(0, positive_positions_max)

# Plot 3: Scatter plot of TP_positions vs total_positions
sns.scatterplot(x="total_positions", y="TP_positions", hue="Model", data=results_sorted, s=100, ax=ax[2, 1])
ax[2, 1].set_title("TP_positions vs Total Positions by Model", fontsize=16)
ax[2, 1].set_xlabel("Total Positions", fontsize=14)
ax[2, 1].set_ylabel("TP Positions", fontsize=14)
total_positions_max = results_sorted["total_positions"].max()
ax[2, 1].set_xlim(0, total_positions_max)
ax[2, 1].set_ylim(0, total_positions_max)

# Improve layout
plt.tight_layout()
plt.show()
```
