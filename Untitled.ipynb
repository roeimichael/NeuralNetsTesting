```python
import pandas as pd

results = pd.read_csv("intermediate_results.csv")
results_sorted = results.sort_values(by=['Avg_ROI', 'Precision','Recall', 'TP_positions'], ascending=False)

# Display top 5 models
print(results_sorted.head(5))

# If you want to visualize these results, seaborn is a useful library
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=results, x='Accuracy', y='Avg_ROI', hue='Optimizer')
plt.show()
```

```python
results = pd.read_csv("intermediate_results.csv")
mask = (results['Avg_ROI'] > 0.02) & (results['Precision'] > 0.25) & (results['total_positions'] >= 10) & (results['total_positions'] <= 350)

filtered_df = results[mask]
results_sorted = filtered_df.sort_values(by=['Avg_ROI', 'Precision','Recall', 'TP_positions'], ascending=False)
top_models_list = []

for idx, row in results_sorted.iterrows():
    model_name = row['Model']
    model_params = row['Model_Params']  # assuming these are stored as dictionary
    model_optimizer = row['Optimizer']
    top_models_list.append({"name": model_name, "params": model_params,"optimizer": model_optimizer})
top_models_list
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=results_sorted, x='Precision', y='Cost_Matrix_Value', hue='Model')
plt.show()
```
